##### Teorema das Raízes Racionais

Seja a<sub>n</sub>x<sup>n</sup>+a<sub>n-1</sub>x<sup>n-1</sup>+...+a<sub>1</sub>x+a<sub>0</sub> uma equação com coeficientes inteiros. Se p/q (fração irredutível) é uma raiz, então p é divisor de a<sub>0</sub> e q é divisor de a<sub>n</sub>.

**OBS:**

1. O teorema não garante a existência de raízes racionais;
2. Encontrada a raiz, use Briot-Ruffini para determinar as demais;
3. Se **a<sub>n</sub>=1** as possíveis raízes são os divisores de **a<sub>0</sub>**.

Em resumo o que temos que fazer é:

1. Encontar possíveis raízes
2. Testar as possíveis raízes até encontrar uma que seja raiz (exemplo P(2)=0)
3. Aplicar Briot-Ruffini para diminuir o grau da equação, até chegar em uma equação de 2º grau

E na aula de equação de 2º grau eu compareci. É só utilizar Teorema de Bháskara.Já aproveito e resolvo também o outro desafio que é escrever o algoritmo para equação de 2º grau. 

A teoria é sempre mais fácil que a prática, mas vamos ao código:

```python
#função para verificar os divisores
def divisores(num):
    #verifica se o número é negativo
    #e transforma em positivo para entrar no for
    if num<0:
        num = num*-1

    listaDivisores=[]
    for i in range(1, num+1):
        if (num % i) == 0:
            listaDivisores.append(int(i))
            listaDivisores.append(int(i*-1))  
    return listaDivisores

#função para testar as possíveis raízes de p/q sustituindo no polinômio
def polinomio(coeficientes, x):
    resultado = 0
    expoente = len(coeficientes)-1
    for coef in coeficientes:
        resultado = resultado + coef * x ** expoente
        expoente=expoente-1
    return resultado
    
def briotRuffini(raiz, coeficientes):
    novosCoeficientes=[]

    novosCoeficientes.append(coeficientes[0])

    for i in range(len(coeficientes)-2):
        j=i+1
        coef = novosCoeficientes[i] * raiz + coeficientes[j]
        novosCoeficientes.append(coef)
        
    return novosCoeficientes

#entrada de dados do usuário
dados = input('Digite os coeficientes da equação separados por vírgula (do maior grau para o menor e A0!=0)')


#alguns coeficientes de equações que usei durante os testes
#dados = '1, 2, -13, -14, 24'
#dados = '1, 2, 3, -22'
#dados = '1, -4, 2, 3'
#dados = '1, -2, -1, 2'
#dados = '1, -1, -7, 1, 6'
#dados = '3, 1, -2, 4'
#dados = '3,-5,2,3,-1'

#Verificar se os coeficientes digitados são inteiros
try:
    #remove espaços em branco digitados
    w = dados.replace(" ","")

    z = list(map(str,w.split(',')))
    coeficientes=[]

    listaA0 = []
    listaAn = []
    possiveisRaizes = []
    resultado=[]
    listaA0 = []
    listaAn = []

    #percorre a lista de dados digitados e verifica se todos são do tipo inteiro
    for i in range(len(z)):
        a = int(z[i])
        coeficientes.append(a)

    #só executa o código se o coeficiente A0 for diferente de 0 e An for igual a 1
    if (coeficientes[-1]!=0) and (coeficientes[0]==1):
        grau = len(coeficientes)-1

        #divisores de a0
        listaA0 = divisores(coeficientes[-1])
        #divisores de an
        listaAn = divisores(coeficientes[0])

        #encontrando p/q
        for i in range(len(listaA0)):
            for j in range(len(listaAn)):
                possiveisRaizes.append(listaA0[i]/listaAn[j])
        
        while  grau >2:
          
            raiz = [] 
            #testar as possíveis raízes
            for x in possiveisRaizes:
                if polinomio(coeficientes,x) == 0:
                    raiz.append(x)
            
            grau = grau-1 
            resultado.append(raiz[0])
            result = []
            result = briotRuffini(raiz[0], coeficientes)
            coeficientes = result

        #fim while

        if(grau ==2):
            a = coeficientes[0]
            b = coeficientes[1]
            c = coeficientes[2]

            delta = b**2 - 4 * a * c
            x1 = (-b + delta**(1/2)) / (2*a)
            x2 = (-b - delta**(1/2)) / (2*a)
            resultado.append(x1)
            resultado.append(x2)
            print('As raízes são:', resultado)
    else:
        print('O coeficiente A0 não pode ser 0 e o An deve ser 1')
#fim try   
  
except ValueError:
    print('Coeficientes digitados não são inteiros')
```
